```python
import numpy as np
import skimage as ski
```

```python
# Since binary file represents only pixel values, information about image shape
# width and height is needed for correct image representation.

def read_bin_img(file_name, img_shape, type = 'uint8'):
    if (type == 'uint8' or type == 'uint16' or type == 'uint32'):
        input_data = np.fromfile(file_name, dtype=type, offset=0)
    else:
        return -1
        
    img_in = np.reshape(input_data, (img_shape[1],img_shape[0]))
    
    return img_in

def write_bin_img(file_name, img_out, type = 'uint8'):
    file_out = open(file_name, 'wb')
    if (type == 'uint8'):
        file_out.write(np.uint8(img_out.flatten()))
    elif (type == 'uint16'):
        file_out.write(np.uint16(img_out.flatten()))
    elif (type == 'uint32'):
        file_out.write(np.uint32(img_out.flatten()))
    else:
        return -1
 
    file_out.close()
```

```python
img = read_bin_img('lena128.bin', [128, 128])
ski.io.imshow(img);
```

```python
img = read_bin_img('lena128neg.bin', [128,128])
ski.io.imshow(img);
```
